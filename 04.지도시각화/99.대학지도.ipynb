```python
import folium
import requests
from urllib.parse import quote
import pandas as pd
```

```python
with open('data/kakaoapikey.txt') as f:
    kakao_key = f.read()
```

```python
base_url = 'https://dapi.kakao.com/v2/local/search/address.json'
addr = '경기도 수원시 장안구 천천로74번길 50 (정자동)'
url = f'{base_url}?query={quote(addr)}'
url
```

```python
header = {'Authorization': f'KakaoAK {kakao_key}'}
```

```python
result = requests.get(url, headers=header).json()
result
```

```python
df = pd.read_csv('data/대학교.csv')
df
```

```python
school_list = '아주대학교,동남보건대학,수원경기대학교,수원성균관대학교,수원여자대학교,용인경희대학교'.split(', ')
school_list
```

```python
lat_list,lng_list = [],[]
for i in df.index:
    url = f'{base_url}?query={quote(df.주소[i])}'
    result = requests.get(url, headers=header).json()
    lat_list.append(float(result['documents'][0]['y']))
    lng_list.append(float(result['documents'][0]['x']))
```

```python
df['위도'] = lat_list
df['경도'] = lng_list
```

```python
df.to_csv('data/대학교.csv', index=False)
pd.read_csv('data/대학교.csv')
```

```python
df = pd.read_csv('data/대학교.csv')
df['color'] = ['red','green','blue','orange','violet','black','purple']
df
```

```python
map = folium.Map(location=[df.위도.mean(), df.경도.mean()], zoom_start=12)
for i in df.index:
    folium.Circle(
        location=[df.위도[i], df.경도[i]],
        popup=folium.Popup(df.주소[i], max_width=200),
        tooltip=df.이름[i],
        radius= 300, color=df.color[i],fill=True
    ).add_to(map)
title_html = '<h3 align= "center" style="font-size:20px">  수원 인근 대학교 </h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
