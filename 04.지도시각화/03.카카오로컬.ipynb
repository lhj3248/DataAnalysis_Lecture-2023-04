```python
import requests
from urllib.parse import quote
import pandas as pd
```

```python
with open('data/kakaoapikey.txt') as f:
    kakao_key = f.read()
base_url = 'https://dapi.kakao.com/v2/local/search/address.json'
addr = '경기도 수원시 영통구 도청로 30 (이의동)'
url = f'{base_url}?query={quote(addr)}'
url
header = {'Authorization': f'KakaoAK {kakao_key}'}
result = requests.get(url, headers=header).json()
result
result['documents'][0].keys()
lat = float(result['documents'][0]['y'])
lng = float(result['documents'][0]['x'])
lat, lng
df = pd.read_csv('data/수원소재관공서.csv')
df
lat_list,lng_list = [],[]
for i in df.index:
    url = f'{base_url}?query={quote(df.주소[i])}'
    result = requests.get(url, headers=header).json()
    lat_list.append(float(result['documents'][0]['y']))
    lng_list.append(float(result['documents'][0]['x']))
df['위도'] = lat_list
df['경도'] = lng_list
```

```python
df.to_csv('data/수원소재관공서2.csv', index=False)
pd.read_csv('data/수원소재관공서2.csv')
```

```python
import folium
from urllib.parse import quote
import pandas as pd
import requests, json
with open('data/roadapikey.txt') as f:
    road_key = f.read()
quote('경기도청')
base_url = 'https://www.juso.go.kr/addrlink/addrLinkApiJsonp.do'
params1 = f'confmKey={road_key}&currentPage=1&countPerPage=10'
params2 = f'keyword={quote("경기도청")}&resultType=json'
url = f'{base_url}?{params1}&{params2}'

result = requests.get(url)
result.status_code
result.text[:200]
res = json.loads(result.text[1:-1])
res['results']['juso'][0]['roadAddr']
gov_list = '경기도청, 장안구청, 권선구청, 팔달구청, 영통구청'.split(', ')
gov_list
road_addr_list = []
for gov in gov_list:
    params2 = f'keyword={quote(gov)}&resultType=json'
    url = f'{base_url}?{params1}&{params2}'
    result = requests.get(url)
    if result.status_code == 200:
        res = json.loads(result.text[1:-1])
        road_addr_list.append(res['results']['juso'][0]['roadAddr'])
    else:
        print(result.status_code)
import pandas as pd

df= pd.DataFrame({
    '이름' : gov_list, '주소': road_addr_list
})
df
df.to_csv('data/수원소재관공서.csv', index=False)
pd.read_csv('data/수원소재관공서.csv')
```
